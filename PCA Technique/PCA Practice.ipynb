```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import  load_breast_cancer
```

```python
data=load_breast_cancer()

data
```

```python
df=pd.DataFrame(data.data)
df
```

```python
df_1=data.data
target =data.target
features=data.feature_names
features
target
```

```python
df_o=pd.DataFrame(df_1,columns=features)
df_o
```

```python
from sklearn.preprocessing import StandardScaler
df_o_sc=StandardScaler().fit_transform(df_o)
df_o_sc=pd.DataFrame(df_o_sc)
df_o_sc
```

```python
 from sklearn.decomposition import PCA
```

```python
pca =PCA(random_state=100)
```

```python
pca_data =pca.fit(df_o_sc)
pca_data
```

```python
pca_data_t=pca.transform(df_o_sc)
pca_data_t
```

```python
print(pca_data.explained_variance_ratio_)
print('\ncum_ratio\n',np.cumsum(pca_data.explained_variance_ratio_))
```

```python
cum_ratio=np.cumsum(pca_data.explained_variance_ratio_)
cum_ratio[cum_ratio<.96]
```

```python
df_pca_subset=pd.DataFrame(pca_data_t[:,0:10],columns=['PC'+str(i) for i in range(10)])
df_pca_subset
```

```python
from sklearn.model_selection import train_test_split as split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report as report
from sklearn.metrics import confusion_matrix as matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score
lr =LogisticRegression()
```

```python
#Data
x=df_pca_subset
y=target

# splitting
x_train,x_test,y_train,y_test=split(x,y,test_size=0.3,random_state=42)

# model fitting
lr.fit(x_train,y_train)

#model predict
train_pred=lr.predict(x_train)
test_pred=lr.predict(x_test)

#result
print('\n tr_report',report(y_train,train_pred))
print('\n ts_report\n',report(y_test,test_pred))    
print('\n tr_matrix\n',matrix(y_train,train_pred))
print('\nts_matrix\n\n ',matrix(y_test,test_pred))
 
```
