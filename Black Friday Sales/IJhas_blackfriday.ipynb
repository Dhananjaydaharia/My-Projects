```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv("train.csv")
```

```python
train_df.shape
```

```python
train_df.head()
```

```python
test_df=pd.read_csv("test.csv")
```

```python
test_df.head()
```

```python
##appending the both dataset
df=train_df.append(test_df)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Changing the Gender into numerical feture
df['Gender']=df['Gender'].map({'F':1,'M':0})
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6, '55+':7})
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
```

```python
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

###  fixing the null values
replacing missing values

```python
 df['Product_Category_1'].unique()
```

```python
df["Product_Category_1"].value_counts()
```

```python
###in descrete feuture the mode is the better way to treate the missing  values
## it change the nan values to mode of the given dataset
df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2']).mode()[0]
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
##in the case of Product_Category_3 do the same
df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3'].mode()[0]
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3']).mode()[0]
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
##we have to change the Stay_In_Current_City_Years object as int, because we changed it into int
```

```python
##convert object into intiger
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
 df['B']=df['B'].astype(int)
```
