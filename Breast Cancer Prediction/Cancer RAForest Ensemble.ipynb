```python
data=pd.read_csv('cancer.csv') 
df=data.copy(deep=True)
df
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
df_null=pd.DataFrame({'count':df.isnull().sum(),'percent':(df.isnull().sum()/df.shape[0])*100})
df_null=df_null.sort_values('percent',ascending =False)
df_null
```

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
df
```

```python
df.diagnosis
```

```python
df['diagnosis']=df['diagnosis'].replace(['M','B'],[0,1])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
from sklearn.preprocessing import StandardScaler ,MinMaxScaler
```

```python
from sklearn.preprocessing import LabelEncoder as lbe
from sklearn.model_selection import train_test_split as split
from sklearn.metrics import  accuracy_score, confusion_matrix,classification_report
```

```python
from  xgboost import XGBClassifier
```

```python
sc=StandardScaler()
msc=MinMaxScaler()
```

```python
y=df['diagnosis']
x=df.drop('diagnosis',axis =1)

x_train,x_test,y_train,y_test =split(x,y,test_size=0.25,random_state =100)
```

```python
# Scaling the data
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.fit_transform(x_test)
```

```python
model=XGBClassifier(objective='reg:logistic',use_label_encoder=False)
```

```python
# model setting with trained data with no scaling
model.fit(x_train,y_train)
```

```python
# test predictions
test_pred=model.predict(x_test)
predictions = [round(value) for value in test_pred]
predictions[:10]
```

```python
# evaluate predictions
accuracy = np.mean(y_test == predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

cm = confusion_matrix(y_test, predictions)
print("\nConfusion Matrix \n",cm)

print('classification report',classification_report(y_test, predictions))
```

# USing Dmatrix

```python

x_tr_d =DMatrix(x_train_scaled,label=y_train)
x_ts_d =DMatrix(x_test_scaled)
```

```python
# setting parameters
import xgboost as xgb
params={'max_depth':5,'objective':'reg:logistic','eta':0.05}

model=xgb.train(params,x_tr_d)
pred =model.predict(x_ts_d)
```

```python
Dpredictions = [round(value) for value in pred]
Dpredictions[:10]
```

```python
# evaluate predictions
accuracy = np.mean(y_test == Dpredictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

cm = confusion_matrix(y_test, Dpredictions)
print("\nConfusion Matrix \n",cm)

print('classification report',classification_report(y_test, Dpredictions))
```
