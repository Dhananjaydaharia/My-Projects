## Gradient Descent Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([0.5,2.3, 2.9, 3.3])
y = np.array([1.4, 1.9, 3.2, 3.8])
```

```python
plt.scatter(x, y)
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```

```python
def gr_des(x,y,int, slp, lr):
    N = len(y)
    loss = int + (slp * x)
    serr = round(np.sum(y - loss)/N,5)
    cur_int = np.sum((y - loss)) * (-2/N)
    cur_slp = np.sum(x * (y - loss)) * (-2/N)
    new_int = int - lr * cur_int
    new_slp = slp - lr * cur_slp
    
    return (new_int, new_slp, serr)
    
```

```python
int = 0
slp = 0
lr = 0.1
i = 1
max_iter = 1000
pint, pslp = 1,1

intlst = slplst = [0]
while ((i < max_iter) and ((pint != int) and (pslp != slp))):
    pint, pslp = int,slp
    int, slp,serr = gr_des(x, y, int, slp, lr)
    print(i,int, slp,serr)
    intlst.append(int)
    slplst.append(slp)
    i += 1

```

```python
plt.scatter(x, y)
plt.plot(x, intlst[-1]+x*slplst[-1])
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```
